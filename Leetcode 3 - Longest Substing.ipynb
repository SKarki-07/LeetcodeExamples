```python
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        i = 0
        j = 1
        current_longest = s [i:j]
        old_longest = s[i:j]
        while ( j < len(s)):
            if (s[j] in current_longest):
                i+=1
                current_longest=s[i:j]
            else:
                j +=1
                current_longest=s[i:j]
                if (len(current_longest)>len(old_longest)):
                    old_longest=current_longest
        return len(old_longest)

s=Solution()
print(s.lengthOfLongestSubstring("abcded"))

            
```
