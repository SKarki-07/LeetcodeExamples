```python
from typing import List
class solution():
    def minSubArrayLen(self, target:int, nums:List[int])->int:
        l  = 0
        cur_sum=0
        min_length=float("inf")

        for r in range(len(nums)):
            cur_sum+=nums[r]

            while cur_sum>=target:
                if r-l+1 < min_length:
                    min_length=r-l+1
                cur_sum-=nums[l]
                l+=1

        return min_length if min_length != float("inf") else 0


```
