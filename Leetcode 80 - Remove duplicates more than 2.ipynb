```python
from typing import List
class Solution:
    def removeDuplicates(self, nums: List[int]) -> int:
        if len(nums)<=2:
            return len(nums)
        
        i = 0
        counter = 1

        for j in range (1, len(nums)):
            if nums[j]!=nums[i]:
                nums[i+1]=nums[j]
                i+=1
                counter=1
            elif nums[j]==nums[i] and counter==1:
                nums[i+1]=nums[j]
                i+=1
                counter+=1
        return i+1
                
```
