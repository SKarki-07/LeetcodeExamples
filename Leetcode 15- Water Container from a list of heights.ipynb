```python
from typing import List

class Solution:
    def maxArea (self, height:List[int])->int:
        
        i=0
        j=len(height)-1
        old_area = 0

        while j>i:
           
            if height[i]-height[j]<0:
                h = height[i]
                new_area=(j-i)*h
                print(new_area, h, (j-i))
                i+=1
                
            else:
                h = height[j]
                new_area=(j-i)*h
                print(new_area, h,(j-i) )
                j-=1
                

            if new_area > old_area:
                old_area = new_area
                
                     


        return old_area

s = Solution()
print(s.maxArea([1,8,6,2,5,4,8,3,7]))



```
