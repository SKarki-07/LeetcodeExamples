```python
class solution:
    def removeDuplicates(self,nums:list[int])->int:
        if len(nums)==0:
            return 0
        
        writepointer = 0
        
        for readpointer in range(1,len(nums)):
            if nums[readpointer]!=nums[readpointer-1]:
                nums[writepointer]=nums[readpointer-1]
                writepointer+=1
        
        for i in range(0,writepointer+1):
            print(nums[i])

        print('break')
        return writepointer+1 #because i don't deal with edge element having +1 makes sense here, just try to repeat
                
    
#call the solution
s = solution()

print(s.removeDuplicates([1,1,1,2,2,3,3,5,6,6,6,8,9,9,10]))

```
